```python
import math

def is_prime(n):
    count=0
    for i in range (2,int(math.sqrt(n)+1)):
        if(n%i == 0):
            count+=1

    if(count<1):
        return True
    else:
        return False

for i in range (1,51):
    if(is_prime(i)):
        print(f"{i} is prime")
```
